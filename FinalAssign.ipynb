After doing an initial analysis on the prices, i will apply the correlation function to get some additional insights on the relationship between each attribute and the target(price). The correlation function gives a values between 1(Strong positive correlation) to -1(Strong negative correlation) with 0 correlation meaning no relationship. It reflects how similar the measurements of variables are across a dataset and identifies patterns and relationships that exist within the data. However, the default .corr() function (which uses pearson's method) is usually applied on numeric variables.  

```python
correlation = df.corr()
price_corr = correlation['price'].sort_values(ascending=False)
print(price_corr)
```

The correlation method provides us with some good insights; there is a strong correlation between the price and the square foot of living space. However concerning the different variable we have; categorical variables and geographical variables, there isn't much to inspect through this method and it doesnt provide good values on such correlation. The df.corr() method works on continuous numerical variables, but not for ranked/geographic variables. The "kendall" method on the other side, does work on ranked variables and gives better results for them.

```python
correlation = df.corr(method='kendall')
price_corr = correlation['price'].sort_values(ascending=False)
print(price_corr)
```

Through this kendall correlation method, the attribute with the highest correlation is the grade, hence we have to focus on it through our analysis to get further insights about it.

Through the above dataframes, we can get many insights.
